```python
import pandas as pd

entries = [["Cowboy Bebop", "TV", "Original", "Action/Adventure/Comedy/Drama/Sci-Fi/Space"]]

df = pd.DataFrame(data=entries, columns=["name", "type",	"source",	"genre"])
df.head()
```

```python
df["genre"] = df["genre"].str.split("/")
df.head()
```

```python
df = df.explode("genre")
df.head() 
```
